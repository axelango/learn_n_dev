```python
#load image
test_image_1 = image.load_img('/content/gdrive/My Drive/input_examples/stop_sign.png', target_size = (128, 128, 1))
test_image_2 = image.load_img('/content/gdrive/My Drive/input_examples/real_stop_sign.jpg', target_size = (128, 128, 1))
```

```python
#convert image to array
test_image_1 = image.img_to_array(test_image_1)
test_image_2 = image.img_to_array(test_image_2)
```

```python
#rgb values to grayscale
test_image_1 = np.mean(test_image_1, axis=2)
test_image_2 = np.mean(test_image_2, axis=2)
```
