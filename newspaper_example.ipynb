# Scraping websites with Newspaper3k in Python

Web Scraping is a powerful tool to gather information from a website. To scrape multiple URLs, we can use a Python library called Newspaper3k. The Newspaper3k package is a Python library used for Web Scraping articles, It is built on top of requests and for parsing lxml. This module is a modified and better version of the Newspaper module which is also used for the same purpose.

Step-by-step Approach:
First we will define a list containing the URLs or assign a single URL.

We will create an Article object passing in the parameters such as the name of the URL and optional parameters     like language=’en’, for English

We will then download and parse the file.
Finally, display the data extracted.

### Example 2

Here, we scrap data from multiple URLs and then display it.

```python
# Import required modules
import newspaper

# Define list of urls
list_of_urls = ['https://www.geeksforgeeks.org/how-to-get-the-magnitude-of-a-vector-in-numpy/',
				'https://www.geeksforgeeks.org/3d-wireframe-plotting-in-python-using-matplotlib/',
				'https://www.geeksforgeeks.org/difference-between-small-data-and-big-data/']

# Parse through each url and display its content
for url in list_of_urls:
	url_i = newspaper.Article(url="%s" % (url), language='en')
	url_i.download()
	url_i.parse()
	print(url_i.text)
```

```python
for article in web_paper.articles:
    print(article.url)
```

```python
for category in web_paper.category_urls():
    print(category)
```

```python
for feed_url in web_paper.feed_urls():
    print(feed_url)
```

**Combine everything:**

```python
import newspaper
from newspaper import Article

import nltk
nltk.download('punkt')

def spider_newspaper_url(url):
    """
     by default, newspaper cache all previously extracted articles and delete any articles it has extracted 。
     using memoize _articles parameter to exit this feature. 
    """
    web_paper = newspaper.build(url, language="zh", memoize_articles=False)
    print(" extract the url of the news page ！！！")
    for article in web_paper.articles:
    #  get the url of the news page 
        print(" news page url :", article.url)
#  call the spiders _newspaper_information function gets news page data 
        spider_newspaper_information(article.url)

    print(" a total of obtaining %s article " % web_paper.size())  #  number of articles 

#  get information about the article 
def spider_newspaper_information(url):
    #  build links and download articles 
    article = Article(url, language='zh')  # Chinese
    article.download()
    article.parse()

#  get information about the article 
    print("title=", article.title)  #  get article title 
    print("author=", article.authors)  #  get article author 
    print("publish_date=", article.publish_date)  #  get article date 
    # print("top_iamge=", article.top_image)  #  get the image address at the top of the article 
    # print("movies=", article.movies)  #  get the article video link 
    print("text=", article.text, "\n")  #  get the text 
    print("summary=", article.summary)  #  get an abstract of the article 


if __name__ == "__main__":
    web_lists = ["http://www.cnn.com/style/article/vermeer-gemaldegalerie-2021-hidden-cupid/","http://www.cnn.com/style/article/tetris-video-game-history/"]
    for web_list in web_lists:
        spider_newspaper_url(web_list)
```
